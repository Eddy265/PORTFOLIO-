```python
import pandas as pd
```

```python
df = pd.read_csv('http://bit.ly/uforeports')
```

```python
df.isnull().sum()
```

```python
df['Shape Reported'].fillna(value='go', inplace=True)
df['City'].fillna(value='come', inplace=True)
```

```python
df.isnull().sum()
```

```python
import plotly.express as px
```

```python
fig = px.choropleth(df,
                    locations = 'State',
                    locationmode = 'USA-states',
                    scope = 'usa',
                    color = 'Shape Reported',
                    hover_name = 'City',
                    hover_data = ['Shape Reported', 'City','Time'],
                    color_continuous_scale = 'armyrose',
                    title = 'Edwin Big head x2')
```

```python
fig.show()
```
